```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
import requests
```

```python
#Extract CSV into DataFrame from movies
movies_df = pd.read_csv("OutputData/movies.csv")
movies_df
```

```python
#Extract CSV into DataFrame
platform_df = pd.read_csv("OutputData/platform.csv")
platform_df
```

```python
#Extract CSV into DataFrame
moviesource_df = pd.read_csv("OutputData/MovieSource.csv")
moviesource_df
```

```python
#Create Database connection
connection_string = "postgres:postgres@localhost:5432/movies_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#Use pandas to Load  csv converted DataFrames into database
movies_df.to_sql(name='moviesandshows', con=engine, if_exists='append', index=False)
```

```python
#confirm data has been added by quering the table
pd.read_sql_query('select * from moviesandshows', con=engine).head()
```

```python
#Use pandas to Load  csv converted DataFrames into database
platform_df.to_sql(name='source_table', con=engine, if_exists='append', index=False)
```

```python
#confirm data has been added by quering the table
pd.read_sql_query('select * from source_table', con=engine).head()
```

```python
#Use pandas to Load  csv converted DataFrames into database
moviesource_df.to_sql(name='moviesource', con=engine, if_exists='append', index=False)
```

```python
#confirm data has been added by quering the table
pd.read_sql_query('select * from moviesource', con=engine).head()
```

```python
pd.read_sql_query("SELECT m.imdb_id, title FROM moviesandshows m JOIN moviesource a on m.imdb_id=a.imdb_id JOIN source_table p on p.platform_id= a.platform_id WHERE platform_name='Netflix'", con=engine)
```
