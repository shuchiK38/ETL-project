```python
import pandas as pd
import numpy as np
import requests
from d_config import api_ombd 
```

```python
# url created
query_url="https://api-v2.crackle.com/browse/all?sortOrder=alpha-asc&pageNumber=1&pageSize=300"
response = requests.get(query_url).json()
response
```

```python
# get various ratings from omdb link 
def get_rating(ratings, type):
    for dic in ratings:
        if dic["Source"] == type:
            return dic["Value"]
    return "None"
```

```python
#crackle movies, lists

IMDB_id = []
title = []
Year = []
Rated = []
Released = []
Runtime = []
Genres = []
Director = [] 
Writer = []
Actors = []
Language = []
Country = []
IMDb_Ratings = []
RT_Ratings = []
Metacritic_Ratings = []
Type_MS = []



i=0
for movie in response["Items"]:
    
    pars = {"t" : movie["Title"], "apikey": api_ombd}
    omdb_resp = requests.get(f'http://www.omdbapi.com/',params=pars).json()
    if omdb_resp["Response"] == "True":
        IMDB_id.append(omdb_resp["imdbID"])
        title.append(omdb_resp["Title"])
        Year.append(omdb_resp["Year"])
        Rated.append(omdb_resp["Rated"])
        Released.append(omdb_resp["Released"])
        Runtime.append(omdb_resp["Runtime"])
        Genres.append(omdb_resp["Genre"])
        Director.append(omdb_resp["Director"])
        Writer.append(omdb_resp["Writer"])
        Actors.append(omdb_resp["Actors"])
        Language.append(omdb_resp["Language"])
        Country.append(omdb_resp["Country"])
        IMDb_Ratings.append(omdb_resp["imdbRating"])
        RT_Ratings.append(get_rating(omdb_resp['Ratings'], "Rotten Tomatoes"))
        Metacritic_Ratings.append(get_rating(omdb_resp['Ratings'],"Metacritic"))
        Type_MS.append(omdb_resp["Type"])    
    else:
        IMDB_id.append('')
        title.append(movie["Title"])
        Year.append(movie["ReleaseYear"])
        Rated.append(movie["Rating"])
        Released.append(movie["ReleaseYear"])
        Runtime.append('')
        Genres.append(", ".join(movie["Genres"]))
        Director.append(", ".join(movie["Directors"]))
        Writer.append(", ".join(movie["Writers"]))
        Actors.append(", ".join(movie["Cast"]))
        Language.append(movie["DefaultLanguage"])
        Country.append(movie["GeoCode"])
        IMDb_Ratings.append('')
        RT_Ratings.append('')
        Metacritic_Ratings.append('')
        if movie["IsEpisodic"] == "True":
           Type_MS.append("series")
        else:
           Type_MS.append("movie")
              
      
    i+=1
    if i%25 == 0:
        print(f'{i},')
#    if i>2:
#        break
        
        
print(title)
#print(Genres)
#print(omdb_resp)
#{'Response': 'False', 'Error': 'Movie not found!'}
```

```python
movies = {"IMDB_id": IMDB_id,"Title" : title,"Year": Year,"Rated" : Rated,
         "Released": Released, "Runtime": Runtime, "Genre": Genres, "Director": Director, 
         "Writer": Writer, "Actors": Actors, "Language" : Language, "Country" : Country, 
          "IMDB_rating": IMDb_Ratings, "RT_rating": RT_Ratings, "Metacritic_rating": Metacritic_Ratings,
          "Type": Type_MS
}

mo=pd.DataFrame(movies)
mo.head(25)
```

```python
movie_NF_df = pd.DataFrame(mo[mo.IMDB_id==''])
movie_NF_df.head()
```

```python
# cleaning the data for various characters and strings to be able finding movie from crackle in omdb by Title (and Year) for imdb_id lookup

movie_NF_df['Title'] = movie_NF_df['Title'].str.replace(r" \(.*\)","")
movie_NF_df['Title'] = movie_NF_df['Title'].str.replace("&","and")
movie_NF_df['Title'] = movie_NF_df['Title'].str.replace("Hairy Bikers'","")
movie_NF_df['Title'] = movie_NF_df['Title'].str.replace("Lindners","")
movie_NF_df['Title'] = movie_NF_df['Title'].str.replace("It's Showtime","Showtime")
movie_NF_df['Title'] = movie_NF_df['Title'].str.replace("William Shatner's","")

```

```python
to_fix_list = movie_NF_df["Title"].to_list()
to_fix_list_y = movie_NF_df["Year"].to_list()

to_fix_list
```

```python
# movies not found with imdb_id - to_fix_list as a list with Title and Year, pushed through loop with omdb api to lookup it again 

IMDB_id = []
title = []
Year = []
Rated = []
Released = []
Runtime = []
Genres = []
Director = [] 
Writer = []
Actors = []
Language = []
Country = []
IMDb_Ratings = []
RT_Ratings = []
Metacritic_Ratings = []
Type_MS = []



i=0
for i in range(0, len(to_fix_list)):
    
    pars = {"t" : to_fix_list[i], "apikey": api_ombd, "y":to_fix_list_y[i]}

    omdb_resp = requests.get(f'http://www.omdbapi.com/',params=pars).json()
   
    if omdb_resp["Response"] == "True":
        IMDB_id.append(omdb_resp["imdbID"])
        title.append(omdb_resp["Title"])
        Year.append(omdb_resp["Year"])
        Rated.append(omdb_resp["Rated"])
        Released.append(omdb_resp["Released"])
        Runtime.append(omdb_resp["Runtime"])
        Genres.append(omdb_resp["Genre"])
        Director.append(omdb_resp["Director"])
        Writer.append(omdb_resp["Writer"])
        Actors.append(omdb_resp["Actors"])
        Language.append(omdb_resp["Language"])
        Country.append(omdb_resp["Country"])
        IMDb_Ratings.append(omdb_resp["imdbRating"])
        RT_Ratings.append(get_rating(omdb_resp['Ratings'], "Rotten Tomatoes"))
        Metacritic_Ratings.append(get_rating(omdb_resp['Ratings'],"Metacritic"))
        Type_MS.append(omdb_resp["Type"])
    i+=1
#    if i>2:
#        break

# creating dataframe with results from omdb lookup              
movies2 = {"IMDB_id": IMDB_id,"Title" : title,"Year": Year,"Rated" : Rated,
         "Released": Released, "Runtime": Runtime, "Genre": Genres, "Director": Director, 
         "Writer": Writer, "Actors": Actors, "Language" : Language, "Country" : Country, 
          "IMDB_rating": IMDb_Ratings, "RT_rating": RT_Ratings, "Metacritic_rating": Metacritic_Ratings,
          "Type": Type_MS
}

mo2=pd.DataFrame(movies2)
```

```python
# dataframe with movies found from second lookup with omdb api 
mo2
```

```python
#dataframe with movies with imdb_id found only (cleaned from movies without imdb_id)
crackle_omdb_df = pd.DataFrame(mo[mo.IMDB_id !=""])
#merging with dataframe with movies from second lookup with imdb_id found after cleaning
crackle_final_df = crackle_omdb_df.append(mo2)
```

```python
crackle_final_df ['Country'] = crackle_final_df ['Country'].str.replace("USA","United States")
crackle_final_df ['Country'] = crackle_final_df ['Country'].str.replace("UK","United Kingdom")
crackle_final_df
```
